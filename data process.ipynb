改檔案名稱

```python
import os
import shutil

# 起始資料夾的路徑
source_folder = './images/full'

# 目標資料夾的路徑
target_folder = './images/process'

# 確保目標資料夾存在，如果不存在則創建
if not os.path.exists(target_folder):
    os.makedirs(target_folder)

# 獲取起始資料夾中的所有檔案
file_list = os.listdir(source_folder)

# 迭代處理每個檔案
for file_name in file_list:
    # 舊的檔案路徑
    old_file_path = os.path.join(source_folder, file_name)

    # 新的檔案名稱，這裡我們將檔名更改為.jpg
    new_file_name = file_name + '.jpg'

    # 新的檔案路徑
    new_file_path = os.path.join(target_folder, new_file_name)

    # 複製檔案
    shutil.copy(old_file_path, new_file_path)

print("檔案複製完成")
```

```python
import json
from collections import defaultdict

# 打開JSON檔案並讀取數據
with open('guru.json', 'r') as json_file:
    data = json.load(json_file)

# 初始化字典以存儲相同數量的attributes的title數量
title_counts_non_empty = defaultdict(int)
title_counts_empty = defaultdict(int)

# 遍歷每個JSON對象
for entry in data:
    # 初始化title的屬性計數為0
    attribute_count = 0

    # 遍歷指定的屬性列表
    attributes = ['general_impression', 'depth_of_field', 'focus', 'color_lighting', 'composition', 'description', 'subject_of_photo', 'use_of_camera']
    
    for attribute in attributes:
        value = entry.get(attribute, '')  # 使用get方法獲取屬性值，如果為空則使用空字符串
        if value and len(value) < 500:
            attribute_count += 1  # 如果屬性值非空，則增加計數

    # 將title的屬性數量加入相應數量的attributes的title數量中
    if attribute_count > 0:
        title_counts_non_empty[attribute_count] += 1
    else:
        title_counts_empty[attribute_count] += 1

# 輸出每個相同數量非空attributes的title數量
for count, title_count in title_counts_non_empty.items():
    print(f"{count} 個非空屬性的title數量： {title_count} 個")

# 輸出每個相同數量空attributes的title數量
for count, title_count in title_counts_empty.items():
    print(f"{count} 個空屬性的title數量： {title_count} 個")
```

```python
import json

# 打開JSON檔案並讀取數據
with open('guru.json', 'r') as json_file:
    data = json.load(json_file)

# 初始化計數器字典
title_attribute_count = {}

# 創建七個不同的TXT檔案，每個屬性一個
attributes = ['general_impression', 'depth_of_field', 'focus', 'color_lighting', 'composition', 'description', 'subject_of_photo', 'use_of_camera']
txt_files = {attribute: f'{attribute}_data.txt' for attribute in attributes}

# 初始化每個屬性的TXT檔案的寫入器
writers = {attribute: open(file, 'w', newline='', encoding='utf-8') for attribute, file in txt_files.items()}
for writer in writers.values():
    writer.write("image,caption\n")  # 寫入標題行

# 遍歷每個JSON對象
for entry in data:
    # 獲取"title"和"image_url"
    title = entry['title'] + '.jpg'

    # 遍歷每個屬性，並寫入對應的TXT文件
    for attribute in attributes:
        value = entry.get(attribute, '')  # 使用get方法獲取屬性值，如果為空則使用空字符串
        if value and len(value) < 500:
            writers[attribute].write(f"{title},{value}\n")
            title_attribute_count.setdefault(title, []).append(attribute)

# 關閉所有TXT檔案
for writer in writers.values():
    writer.close()

# 輸出每個title的attribute數量
for title, attributes_list in title_attribute_count.items():
    print(f"{title} 有 {len(attributes_list)} 個attribute。")

print("轉換完成")
```
